```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences 
sent=['the glass of milk', 'the glass of juice', 
      'the cup of tea', 'I am a good boy', 'I am a good developer',
      'understand the meaning of words', 'I would like to travel'] 
```

```python
sent
```

1. one hot
2. pre/post purnig (convert sent to same length)
3. Invoking Emdedding layer (same as word2vec)

```python
# Vocabulary Size --- No of unique words in corpus
voc_size = 500
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
one_hot_resp=[one_hot(words,voc_size) for words in sent]    # Find Index Positions
one_hot_resp 
```

```python
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.preprocessing.sequence import pad_sequences  #zero padding
from tensorflow.keras.models import Sequential 
```

```python
sent_length = 8  #length of sentence
embedded_docs = pad_sequences(one_hot_resp,padding='pre',maxlen=sent_length) # padding = 'pre'/'post' 
embedded_docs
```

```python
dim=10 
model= Sequential() 
model.add(Embedding(voc_size,dim,input_length=sent_length))   #dim -- feature dimension
model.compile('adam','mse') 
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])  # Convert each words to word2vec having 10 dimension/features
```
